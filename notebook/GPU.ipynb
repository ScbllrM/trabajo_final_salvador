```python
import torch
```

```python
print(f"___PyTorch_____________")
print(f"Versión CUDA:     {torch.version.cuda}")
print(f"Versión cuDNN:    {torch.backends.cudnn.version()}")
print(f"cuDNN habilitado: {torch.backends.cudnn.enabled}")
print(f"Versión PyTorch:  {torch.__version__}")
print("")

if torch.cuda.is_available():
    print(f"____Información GPU___________")
    print(f"N° GPUs: {torch.cuda.device_count()}")
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    prop = torch.cuda.get_device_properties(0)
    memoria_total = prop.total_memory / 1024**3
    print(f"Capacidad: {memoria_total:.2f} GB")
    print(f"Multiprocesos {prop.multi_processor_count}")
```
